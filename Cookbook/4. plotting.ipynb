```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.plot(np.arange(10), np.arange(10)) #plotting a straight line with 2 numpy arrays of same length
```

```python
plt.plot(np.arange(20), np.arange(20)) #just trying, they must have same length
```

```python
plt.plot(np.arange(10), np.exp(np.arange(10))) #plotting exponential
```

```python
plt.figure()
plt.subplot(121)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(122)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
#after subplot there are 3 numbers: the first 2 are NxN and the last are 2 for vertical and 1 for orizzontal
```

```python
plt.figure() 
plt.subplot(221) 
plt.plot(np.arange(10), np.exp(np.arange(10))) 
plt.subplot(222) 
plt.scatter(np.arange(10), np.exp(np.arange(10))) 
plt.subplot(223) 
plt.scatter(np.arange(10), np.exp(np.arange(10))) 
plt.subplot(224) 
plt.scatter(np.arange(10), np.exp(np.arange(10)))
#1 to 4 because there are 4 plots
```

```python
#FINAL TRY WITH THE REAL DATA
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
target = iris.target

plt.figure(figsize=(12,5)) #resize figure
plt.subplot(121)
plt.scatter(data[:,0], data[:,1], c=target) #first 2 columns
plt.subplot(122)
plt.scatter(data[:,2], data[:,3], c=target) #last 2 columns
#target takes array of colours 0,1,2
```
